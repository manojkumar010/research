## Comparison with feature selection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold
import mrmr
import numpy as np
```

```python

# Load your data
datap=r"C:\Users\kumar\OneDrive\Desktop\Programs\Research@AIIMS\KKD AFT_SCA AI-ML_Calculate.xlsx"
df = pd.read_excel(datap)
df.tail()
```

```python
df.info()
df.describe().T
```

```python

# Define label
df["SCA_type"] = df["Groups"].map({1: "SCA1", 2: "SCA2"})

# Define 19 AFT features based on dictionary
aft_features = [
    'RRI_SDNN', 'RRI_RMSSD', 'RRI_LF_nu', 'RRI_HF_nu', 'RRI_Power_nu',
    'RRI_SD1_nu', 'RRI_SD2_nu',
    'SBP_SDNN', 'SBP_RMSSD', 'SBP_LF_nu', 'SBP_HF_nu', 'SBP_Power_nu',
    'SBP_All_BRS', 'SBP_αLF', 'SBP_αHF',
    'Sudomotor_Score', 'Cardiovagal_Score', 'Adrenergic_Score', 'CASS'
]

X = df[aft_features]
y = df["SCA_type"]
```

```python
X.head()
```

```python

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.15, random_state=42)

```

```python
# Baseline LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred_all = lda.predict(X_test)
print(">>> Baseline: Using All 19 Features <<<")
print("Accuracy:", accuracy_score(y_test, y_pred_all))
print("Classification Report:\n", classification_report(y_test, y_pred_all))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_all))
```

```python

# Define label
df["SCA_type"] = df["Groups"].map({1: "SCA1", 2: "SCA2"})

# Define 24 AFT features based on dictionary
aft_features = [
    'RRI_SDNN', 'RRI_RMSSD', 'RRI_LF_nu', 'RRI_HF_nu', 'RRI_Power_nu', 'RRI_SD1_nu', 'RRI_SD2_nu', 'SBP_SDNN', 'SBP_RMSSD', 'SBP_LF_nu', 'SBP_HF_nu', 'SBP_Power_nu', 'SBP_All_BRS', 'SBP_αLF', 'SBP_αHF', 'Sudomotor_Score', 'Cardiovagal_Score', 'Adrenergic_Score', 'CASS', 'Age', 'Disease_durn', 'Age of onset', 'ICARS'
]

X = df[aft_features]
y = df["SCA_type"]
X.head()
```

```python
y.head()
```

```python

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.15, random_state=42)

```

```python
# Baseline LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred_all = lda.predict(X_test)
print(">>> Baseline: Using All 24 Features <<<")
print("Accuracy:", accuracy_score(y_test, y_pred_all))
print("Classification Report:\n", classification_report(y_test, y_pred_all))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_all))
```

```python
# Encode target
df_mrmr = df[aft_features + ["SCA_type"]].copy()
df_mrmr["SCA_type"] = LabelEncoder().fit_transform(df_mrmr["SCA_type"])
X_all = df_mrmr.drop(columns=["SCA_type"])
y_all = df_mrmr["SCA_type"]
```

```python
# select top 10 features using mRMR
from mrmr import mrmr_classif
selected_features = mrmr_classif(X=X_all, y=y_all, K=8,return_scores=True)
```

```python
print(selected_features)
```

```python
# Pnew data with selected features
X_selected = df[selected_features]

# Train/test split
X_train_sel, X_test_sel, y_train_sel, y_test_sel = train_test_split(X_selected, y_encoded, test_size=0.15, random_state=42)
```

```python

# Train LDA again
lda_sel = LinearDiscriminantAnalysis()
lda_sel.fit(X_train_sel, y_train_sel)
y_pred_sel = lda_sel.predict(X_test_sel)
print("\n>>> After mRMR Feature Selection <<<")
print("Accuracy:", accuracy_score(y_test_sel, y_pred_sel))
print("Classification Report:\n", classification_report(y_test_sel, y_pred_sel))
print("Confusion Matrix:\n", confusion_matrix(y_test_sel, y_pred_sel))
```
